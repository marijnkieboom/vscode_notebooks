```python
import nltk
from nltk.tokenize import word_tokenize

# Sample text
text = "Ik ga naar de supermarkt, daarna ga ik naar school."

# Tokenize the text into words
tokens = word_tokenize(text)

vocab = set(tokens)

print(vocab)
```

```python
import torch
import spacy
import random
import numpy as np
import torch.nn as nn

# Load the SpaCy model
nlp = spacy.load("nl_core_news_md")

print('len of vocabulary:', len(nlp.vocab.strings))
ix = random.randint(0, len(nlp.vocab.strings) - 1)
print('random vocab entry:', list(nlp.vocab.strings)[ix])

print()

def word_to_vector(word):
    hash = nlp.vocab.strings[word]
    vect = nlp.vocab.vectors[hash]
    return np.asarray([vect])

ms = nlp.vocab.vectors.most_similar(word_to_vector('school'), n=10)

words = [nlp.vocab.strings[w] for w in ms[0][0]]
print(words)

print()

# Example sentence
text = "Dit is een voorbeeld zin. Dit is een andere voorbeeld zin"

# Tokenize the text using SpaCy
doc = nlp(text)

# Initialize nn.Embedding layer with SpaCy embeddings
embedding_layer = nn.Embedding.from_pretrained(torch.FloatTensor(nlp.vocab.vectors.data))

print('emb parameter count:', sum(len(p) for p in embedding_layer.parameters()))

key = nlp.vocab['kat'].orth
print(key)
```

```python
import torch
import spacy
import torch.nn as nn

# Load the SpaCy model
nlp = spacy.load("nl_core_news_sm")

print('len of vocabulary:', len(nlp.vocab.strings))
ix = random.randint(0, len(nlp.vocab.strings) - 1)
print('random vocab entry:', list(nlp.vocab.strings)[ix])

embedding_dim = 300
vocab_size = len(nlp.vocab.strings)
print(vocab_size)

embedding_layer = nn.Embedding(vocab_size, embedding_dim)

print('emb parameter count:', sum(len(p) for p in embedding_layer.parameters()))

```
